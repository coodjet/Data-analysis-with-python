```python
import pandas_datareader as pdr
import datetime as dt
tickers = ['AAPL', 'MSFT', 'NFLX', 'AMZN']
start = dt.datetime(2010, 1, 1)
data = pdr.get_data_yahoo(tickers, start)
data.head()
```

```python
#normalize and calculate return of portfolio
```

```python
data = data['Adj Close']
data.head()
```

```python
# we need to normalize data

norm = data/data.iloc[0]
norm.head()

```

```python

norm.head()
```

```python
# lets assume that our portfolio consists of 25% of each sticker
```

```python
portfolio = [.25, .25, .25, .25]
weights = (norm*portfolio)
weights.head()
```

```python
weights['Total'] = (norm*portfolio).sum(axis=1)
weights.head()
```

```python
# lets assume we invested 100,000 usd
(weights*100000).head()
```

```python
# lets assume we invested 100,000 usd
(weights*100000).tail()
```
